```python
#%matplotlib inline             # просмотр графиков в браузере
import numpy as np
# функция просмотра чтения и сохранения
from skimage.io import imread, imshow, imsave
```

```python
img = imread ('tiger-border.png')
imshow(img)
```

```python
color = img [0, 0]
left = 0
up = 0
right = 0
down = 0

while (img[int(img.shape[0]/2), left, 0] == color[0] and 
       img[int(img.shape[0]/2), left, 1] == color[1] and 
       img[int(img.shape[0]/2), left, 2] == color[2]):
    left=left+1
    
while (img[up, int(img.shape[1]/2), 0] == color[0] and 
       img[up, int(img.shape[1]/2), 1] == color[1] and 
       img[up, int(img.shape[1]/2), 2] == color[2] ):
    up=up+1

while (img[int(img.shape[0]/2), int(img.shape[1])-right-1, 0] == color[0] and 
       img[int(img.shape[0]/2), int(img.shape[1])-right-1, 1] == color[1] and 
       img[int(img.shape[0]/2), int(img.shape[1])-right-1, 2] == color[2]):
    right=right+1
    
while (img[int(img.shape[0])-down-1, int(img.shape[1]/2), 0] == color[0] and 
       img[int(img.shape[0])-down-1, int(img.shape[1]/2), 1] == color[1] and 
       img[int(img.shape[0])-down-1, int(img.shape[1]/2), 2] == color[2] ):
    down=down+1
```
