```python
import numpy as numpy
from skimage.io import imread, imshow, imsave
import skimage as skimage
from skimage import img_as_ubyte
from skimage import img_as_float
from numpy import histogram
img = imread ('railroad.png')
imshow(img)
```

```python
img = img_as_float (img)
R = img [:,:,0]
G = img [:,:,1]
B = img [:,:,2]
R_avr = numpy.sum (R/ numpy.size(R))
G_avr = numpy.sum (G/ numpy.size(G))
B_avr = numpy.sum (B/ numpy.size(B))
AVR = (R_avr+G_avr+B_avr)/3
Rw = R_avr/AVR
Gw = G_avr/AVR
Bw = B_avr/AVR
R = numpy.clip (R/Rw, 0, 1)
G = numpy.clip (G/Gw, 0, 1)
B = numpy.clip (B/Bw, 0, 1)
new_img = img_as_ubyte ( numpy.dstack ((R, G, B)))
imsave ('out_img.png', new_img)
imshow(new_img)

```
