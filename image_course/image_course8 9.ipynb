```python
#%matplotlib inline             # просмотр графиков в браузере
import numpy as numpy
# функция просмотра чтения и сохранения
from skimage.io import imread, imshow, imsave
import skimage as skimage
from skimage import img_as_ubyte
img = imread ('tiger-low-contrast.png')
imshow(img)
```

```python
from matplotlib.pyplot import hist
values, bin_edges, patches = hist(img.ravel(), bins=range(257))
```

```python
new_img = (img - img.min())*(255/(img.max()-img.min()))
# new_img = img_as_ubyte(new_img)
new_img = new_img.astype('uint8')
imshow(new_img)
```

```python
from numpy import histogram
img = img.astype('float')
k = round (img.size*0.05)
h, bin_edges = histogram(img, bins=range(257))
count = 0
for i in range (255):
    count += h[i]
    if count > k:
        x_min = i
        break

count = 0
for i in range (255,1, -1):
    count += h[i]
    if count > k:
        x_max = i
        break
print (x_min, x_max)
```

```python
new_img = (img - x_min)*(255/(x_max - x_min))
new_img = numpy.clip (new_img, 0, 255)
new_img = new_img.astype('uint8')
# 
imshow(new_img)
imsave ('out_img.png', new_img)
```
